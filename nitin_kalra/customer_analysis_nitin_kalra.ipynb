#1. Find out the avg orders and avg totalsales per customers?

```python
avg =df1[df1['orders']==1]
```

```python
total_order = avg.groupby('customer_id')['orders'].sum().to_frame().reset_index()
```

```python
total_order['avg_order']= total_order['orders'].mean()
total_order.head(3)
```

```python
avg_sale = avg.groupby('customer_id')['total_sales'].mean().to_frame().reset_index()
avg_sale.rename(columns = {'total_sales':'avg_tot_sales'}, inplace = True)
avg_sale.head(3)
```

```python
#avg orders and avg totalsales per customers
avg_sales_ord = avg_sale.merge(total_order)
avg_order_total= avg_sales_ord[['customer_id','avg_order','avg_tot_sales']] 
avg_order_total
```

#2.Find out the customer first and last purchase to identity the no of days the customer is with us?

```python
max = df1.groupby('customer_id')['hour'].max().to_frame().reset_index()
max
```

```python
max.rename(columns = {'hour':'max_hour'}, inplace = True)
```

```python
min = df1.groupby('customer_id')['hour'].min().to_frame().reset_index()
min
```

```python
min.rename(columns = {'hour':'min_hour'}, inplace = True)
```

```python
result_q =max.merge(min)
result_q.head()
```

```python
result_q['days']= max['max_hour']-min['min_hour']
```

```python
#number of days the customer is with us
result_q
```

#3. Find out the Customer purchase frequency?


```python
#selecting non returned order only
order_sele = df1[df1['orders']==1]
order_sele.head(3)
```

```python
order_sele.drop_duplicates()
order_frequency =order_sele.groupby(by=['customer_id'], as_index=False)['order_id'].count()
order_frequency.rename(columns = {'order_id':'frequency'}, inplace = True)
order_frequency
```

#4.What is the highest and least purchases done by the customers?

```python
#highest and least purchases done by the customers
max_sales =df1.groupby('customer_id')['total_sales'].max().to_frame().reset_index()
max_sales.rename(columns = {'total_sales':'max_purchase'}, inplace = True)
min_sales= df1.groupby('customer_id')['total_sales'].min().to_frame().reset_index()
min_sales.rename(columns = {'total_sales':'min_purchase'}, inplace = True)
max_min_purchase =max_sales.merge(min_sales)
max_min_purchase
```
