```python
df_no_na.drop_duplicates(subset='order_id',keep='first',inplace=True)
```

```python
acc=[]
for i in df_no_na.customer_id.unique():
    annual_frequency=df_no_na[df_no_na.customer_id==i].groupby('year')['orders'].count().to_frame()['orders'].mean()
    monthly_frequency=df_no_na[df_no_na.customer_id==i].groupby('month')['orders'].count().to_frame()['orders'].mean()
    weekly_frequency=df_no_na[df_no_na.customer_id==i].groupby('weekday')['orders'].count().to_frame()['orders'].mean()
    acc.append([i,annual_frequency,monthly_frequency,weekly_frequency])
```

```python
purchase_frequency=pd.DataFrame(acc,columns=['customer_id','annual_frequency','monthly_frequency','weekly_frequency'])
```
